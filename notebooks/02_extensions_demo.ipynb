# ggplotpy v0.5 — extension ecosystem demo

Demonstrates lazy `ggplotpy.ext.*` reflection, ggrepel (when installed), and patchwork composition via `PlotComposition`.

```python
from ggplotpy import *
import pandas as pd
from ggplotpy.core.patchwork import from_plot

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/mtcars.csv')
p = (ggplot(df)
     + aes(x='wt', y='mpg', color='factor(cyl)')
     + geom_point()
     + theme_minimal())
p
```

## ggrepel (optional)

Skipped automatically when the R `ggrepel` package is not installed.

```python
try:
    import ggplotpy.ext as ext
    ggrepel = ext.ggrepel
    p_repel = p + ggrepel.geom_text_repel(aes(label='name'))
    p_repel
except AttributeError as exc:
    print('ggrepel not available — install.packages("ggrepel") in R')
    print(exc)
```

## patchwork (optional)

Use `from_plot()` to wrap ggplot objects, then compose with `|` (side-by-side) or `/` (stacked).

```python
try:
    p2 = (ggplot(df)
          + aes(x='hp', y='mpg')
          + geom_point()
          + theme_minimal())
    dashboard = from_plot(p) | from_plot(p2)
    dashboard
except Exception as exc:
    print('patchwork not available — install.packages("patchwork") in R')
    print(exc)
```
