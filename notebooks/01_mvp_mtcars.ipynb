# ggplotpy MVP — mtcars scatter

Minimal notebook demonstrating real ggplot2 from Python.

```python
from ggplotpy import *
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/mtcars.csv')
p = (ggplot(df)
     + aes(x='wt', y='mpg', color='factor(cyl)')
     + geom_point()
     + theme_minimal())
p
```
